### ROUTER Python 

```python
stats = []
car_sockets = []
i = 0
while(True):
    message = socket.recv_multipart()  # Receive the message from OMNeT++
    print(f"Received {message}")
    car_idx = int(re.split('(\d+)', message[0].decode())[1])
    if message[-1].decode() == 'UPDATE_REQ':
        # A car has to know what to do further
        msg_py = np.random.choice(['NN_'+str(car_idx), 'NN_'+str(car_idx), 'NN_'+str(car_idx), 'CON'])
        print(f'Sending {msg_py} \n')
        socket.send_multipart([car_sockets[car_idx], msg_py.encode(), b'END'])
        if msg_py == 'ENDSIM':
            break
    if message[0].decode().startswith('node') and i < 5:  # Hard-coded number of vehicles
        # Sockets' identities
        car_sockets.append(message[0])
        print(f'Sending NN Pars of node {message[0]} \n')
        socket.send_multipart([car_sockets[car_idx], ('NN_'+str(car_idx)).encode(), b'END'])
    if message[-1].decode() == 'STATS':
        # Transmission statistics arrived
        stats.append([message[0]] + message[1:])
        continue

    i+=1
socket.close()
context.term()
```

```python
def parse_if_number(s):
    try: return float(s)
    except: return True if s=="true" else False if s=="false" else s if s else None

def parse_ndarray(s):
    return np.fromstring(s, sep=' ') if s else None
```

```python
# Read the generated .csv file
eight = pd.read_csv(osp.join(path_res, 'eight.csv'), converters = {
    # cast the strings to floats, ndarrays and bool values
    'attrvalue': parse_if_number,
    'binedges': parse_ndarray,
    'binvalues': parse_ndarray,
    'vectime': parse_ndarray,
    'vecvalue': parse_ndarray})
```
